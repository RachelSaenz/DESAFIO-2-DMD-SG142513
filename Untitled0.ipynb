```python
import matplotlib.pyplot as plt

categorias = [
    "Tiendas y ocio",
    "Supermercados y farmacias",
    "Parques",
    "Estaciones de transporte",
    "Lugares de trabajo",
    "Zonas residenciales"
]

valores = [47, 61, 10, 34, 33, 1]

plt.figure(figsize=(10, 6))
plt.barh(categorias, valores, color='skyblue')
plt.xlabel('Cambio porcentual vs referencia')
plt.title('Movilidad en San Salvador - 15 Oct 2022')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig('movilidad_san_salvador_barras.png')  # Guardar imagen
plt.show()
```
